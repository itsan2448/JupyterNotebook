```python
# method overloading
from multipledispatch import dispatch

@dispatch(int,int)
def A(f,s):
    r=f*s
    return r
@dispatch(float,float)
def A(f,s):
    r=f*s
    return r
@dispatch(float,float,float)
def A(f,s,t):
    r=f*s*t
    return r
@dispatch(int,int,int)
def A(f,s,t):
    r=f*s*t
    return r
    
print(A(3.4,4.5,1.1))
print(A(3,4,2))
print(A(3.4,4.13))
print(A(3,4))
```

```python
from multipledispatch import dispatch

@dispatch(int)
def prime(n):
    if n<=1:
        return 0
    if n==2:
        return 1 
    for i in range (2,n):
        if n%i==0:
            return 0
    return 1;
@dispatch(int,int)
def prime(a,b):
    L=[]
    for i in range(a,b):
        if prime(i):
            L.append(i)
    return L
print(prime(-10,100)) 
```

```python
# tictactoe game
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from tabulate import tabulate

class App:
    def __init__(self):  
        self.root = Tk()
        self.root.geometry('400x300')
        self.root.title('Tic Tac Toe')
        self.mark='X'
        self.b = [[0,0,0],[0,0,0],[0,0,0]]
        self.draw()
        self.root.mainloop()

    
    def draw(self):        
        x = 0
        for i in range(3):
            for j in range(3):                
                self.b[i][j]=Button(self.root, text=str(x+1),height=3, width=7, activebackground="blue", activeforeground="white",
                       bg="lightgray", cursor="hand2", overrelief="raised", highlightbackground="black",
                       highlightcolor="green", font=('century gothic',14), command=lambda i=i, j=j: self.click(i, j))
                self.b[i][j].grid(row=i, column=j, sticky='nsew')
                x += 1
        
    def click(self,i,j):        

        if isinstance(self.b[i][j]['text'],str):
            self.b[i][j]['text'] = self.mark
            self.b[i][j].config(state=DISABLED)  # Disabling BUTTON
            if self.won(i, j):
                winner = self.mark + " wins!"
                # print(winner)                
                messagebox.showinfo("winner",f"{self.mark} wins!!")
                
                
            self.mark = 'X' if self.mark == 'O' else 'O'

    def won(self,i,j):
       

        val=self.b[i][j]['text']        
        # all function
        if all(self.b[i][c]['text'] == val for c in range(3)):
            return True        
        if all(self.b[r][j]['text'] == val for r in range(3)):
            return True
        # tl to br
        if i==j:
            if all(self.b[x][x]['text'] == val for x in range(3)):
                return True
        # lb to tr 
        if i + j == 2: 
            if all(self.b[x][2-x]['text'] == val for x in range(3)):
                return True
        return False

obj = App()
```

```python
#linked lists
class Node:
    def __init__(self,data):        
        self.data=data
        self.next=None
        
class LL:
    def __init__(self):
        self.head=None
        
    def add_at_end(self,data):
        tmp=Node(data)
        if self.head==None:
            self.head=tmp
            return 
        else :
            i=self.head
            while(i.next!=None):
                i=i.next
            i.next=tmp
            return 
            
    def print(self):
        i=self.head
        while i is not None:
            print(i.data,end=" ")
            i=i.next
            
    def reverse(self):
        if self.head==None :
            return self.head
        i=self.head
        p=None 
        k=None
        while i is not None:
            k=i.next
            i.next=p
            p=i
            i=k
        self.head=p
        
    def swap_be(self):
        i=self.head
        j=None
        while i.next is not None:
            j=i
            i=i.next
        i.next=self.head.next
        j.next=self.head   
        self.head.next=None
        self.head=i
        
    def remove(self,k):
        if k==1:
            self.head=self.head.next
            return
        c=1
        i=self.head        
        j=None
        while c<k and i.next is not None:
            j=i
            i=i.next
            c+=1
        if i.next is not None:
            j.next=i.next
        
a=LL()
a.add_at_end(5)
a.add_at_end(15)
a.add_at_end(25)
a.add_at_end(35)
a.add_at_end(45)
a.add_at_end(65)
a.add_at_end(95)
a.print()
print()
a.reverse()
a.print()
a.reverse()
print()
print("re-reversed")
a.swap_be()
# print()
a.print()
a.remove(3)
print()
a.print()
a.remove(5)
print()
a.print()
```

```python
# BT
class Node:
    def __init__(self,data):        
        self.data=data
        self.left=None
        self.right=None
        
class BT:
    def __init__(self,data):
        self.root=Node(data)        
    
     
    def preorder(self,node):
        if node is None:
            return 
        print(node.data,end=" ")
        self.preorder(node.left)
        self.preorder(node.right)
    def start_pre(self):
        self.preorder(self.root)

    
    def inorder(self,node):
        if node is None:
            return 
        self.inorder(node.left)
        print(node.data,end=" ")        
        self.inorder(node.right)
    def start_in(self):
        self.inorder(self.root)

    

    def postorder(self,node):
        if node is None:
            return 
        self.postorder(node.left)
        self.postorder(node.right)
        print(node.data,end=" ")        
        
    def start_post(self):
        self.postorder(self.root)

    
    def lev_order(self,node):
        if node is None:
            return 
        q=[]
        q.append(node)
        while(len(q)>0):
            
            node=q.pop(0)
            print(node.data,end=" ")
            if node.left is not None:
                q.append(node.left)
            if node.right is not None:
                q.append(node.right)
              
        
    def start_lev(self):
        self.lev_order(self.root)

    
a=BT(5)
a.root.right=Node(4)
a.root.right.left=Node(2)
a.root.right.right=Node(6)

a.root.left=Node(10)
a.root.left.left=Node(12)
a.root.left.left.left=Node(13)
a.root.left.right=Node(24)

print("pre:")
a.start_pre()
print()
print("in:")
a.start_in()
print()
print("post:")
a.start_post()
print()
print("level order:")
a.start_lev()
```
