```python
def is_prime(n):
    if n<=1:
        return False
    if n==2:
        return True
    else:
        for i in range(2,n-1):
            if n%i==0:
                return False
        return True
```

```python
from tkinter import *
from tkinter import messagebox
import cv2
from PIL import Image, ImageTk

class App:
    def __init__(self):
        self.root = Tk()
        self.root.geometry('4000x4000')
        self.root.title('APPLICATION FORM')
        self.setup_ui()
        self.capture = None
        self.root.mainloop()

    def setup_ui(self):
        labels = ['ID:', 'Name:', 'Picture:']
        for i, label_text in enumerate(labels):
            Label(self.root, text=label_text, font=('century gothic', 14)).grid(row=i, column=0, sticky='w', padx=10, pady=10)

        self.t1 = Entry(self.root)
        self.t1.grid(row=0, column=1, padx=10, pady=10)
        self.t2 = Entry(self.root)
        self.t2.grid(row=1, column=1, padx=10, pady=10)

        self.label_widget = Label(self.root)
        self.label_widget.grid(row=2, column=1, padx=10, pady=10)

        button1 = Button(self.root, text="Open Camera", command=self.open_camera)
        button1.grid(row=2, column=0, padx=10, pady=10)

        button2 = Button(self.root, text="Add More", command=self.save)
        button2.grid(row=2, column=1, padx=10, pady=10)

        button3 = Button(self.root, text="shutdown", command=self.close_cam)
        button3.grid(row=2, column=2, padx=10, pady=10)
        

    def open_camera(self):
        if self.capture is None:
            self.capture = cv2.VideoCapture(0)

        _, frame = self.capture.read() 
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
        img = Image.fromarray(frame)
        photo_image = ImageTk.PhotoImage(image=img)
        
        self.label_widget.configure(image=photo_image)
        self.label_widget.image = photo_image 
        self.label_widget.after(10, self.open_camera)

    def save(self):
        id_value = self.t1.get()
        name_value = self.t2.get()
        if id_value.strip() == '' or name_value.strip() == '':
            messagebox.showerror("Error", "empty cells")
        else:
            messagebox.showinfo("Success", f"ID: {id_value}, Name: {name_value} saved successfully.")
            self.clear_entries()

    def clear_entries(self):
        self.t1.delete(0, END)
        self.t2.delete(0, END)
        # self.close_cam()
        self.__del__

    def __del__(self):
        if self.capture is not None:
            self.capture.release()
    def close_cam(self):
        if self.capture is not None:
            self.capture.release()  # Release the camera when the window is closed
        self.root.destroy()

app = App()
```
